```python
#OOP API
from pycaret.regression import RegressionExperiment
s = RegressionExperiment()
s.setup(air_data, target = 'PM25', session_id = 123)
```

```python
# OOP API
best_oop = s.compare_models()
```

```python
print(best_oop)
```

```python
# OOP API
s.evaluate_model(best_oop)
```

```python
# OOP API
s.plot_model(best_oop, plot = 'residuals')
```

```python
# oop API
plot_model(best_oop, plot = 'feature')
```

```python
# OOP API
s.predict_model(best_oop)
```

```python
# OOP API
predictions = s.predict_model(best_oop, data=air_data)
predictions.head()
```

```python
# OOP API
s.save_model(best_oop, 'my_best_pipeline')
```
